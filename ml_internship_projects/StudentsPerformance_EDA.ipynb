```python
#analysing the data by creating graphs to check the maximum scores of each subject


max_math=df["math score"].max()
max_reading=df["reading score"].max()
max_writing=df["writing score"].max()

y=np.array([max_math,max_reading,max_writing])
x=np.array(["Math","Reading","Writing"])

mpl.title("Maximum Scores",size=15)
mpl.xlabel("Subject",size=10)
mpl.ylabel("Marks",size=10)
mpl.bar(x,y,width=0.4)
mpl.show()
```

```python
#analysing the data by creating graphs to check the minimum scores of each subject


min_math=df["math score"].min() 
min_reading=df["reading score"].min()
min_writing=df["writing score"].min()

y=np.array([min_math,min_reading,min_writing])
x=np.array(["Math","Reading","Writing"])

mpl.title("Minimum Scores",size=15)
mpl.xlabel("Subject",size=10)
mpl.ylabel("Marks",size=10)
mpl.bar(x,y,width=0.4)
mpl.show()
```

```python
#plotting graphs to check avg scores by gender for each subject


avgscore=studdata.groupby("gender")[["math score","reading score","writing score"]].mean()
subs=["Math","Reading","Writing"]
w=0.3
x=np.arange(len(subs))
#print(x)
#print(avgscore.loc["male"])
#print(avgscore.loc["female"])
mpl.figure(figsize=(9, 7))
mpl.bar(x-w/2,avgscore.loc["male"],width=w,label="Male")
mpl.bar(x+w/2,avgscore.loc["female"],width=w,label="Female")
mpl.title("Average Scores by Gender",size=15)
mpl.xlabel("Subjects",size=10)
mpl.ylabel("Scores",size=10)
mpl.xticks(ticks=x,labels=subs,size=10)
mpl.legend(title="Gender",loc="upper right")
mpl.show()
```

```python
#plotting graphs to check avg scores by all groups for each subject



avg_score=studdata.groupby("race/ethnicity")[["math score","reading score","writing score"]].mean()
subs=["Math","Reading","Writing"]
w=0.1
x=np.arange(len(subs))
mpl.figure(figsize=(9,7))
#print(avg_score)
mpl.bar(x+0.1,avg_score.loc["group A"],width=w,label="Group A")
mpl.bar(x+0.2,avg_score.loc["group B"],width=w,label="Group B")
mpl.bar(x+0.3,avg_score.loc["group C"],width=w,label="Group C")
mpl.bar(x+0.4,avg_score.loc["group D"],width=w,label="Group D")
mpl.bar(x+0.5,avg_score.loc["group E"],width=w,label="Group E")
mpl.title("Average Scores by Group")
mpl.xlabel("Subjects")
mpl.ylabel("Scores")
mpl.xticks(ticks=x+0.3,labels=subs)
mpl.legend(title="Groups",loc="lower left")
mpl.show()
```

```python
#plotting graph to check if the test preparation course really helped the students perform better


avg=studdata.groupby("test preparation course")[["math score","reading score","writing score"]].mean()
subs=["Math","Reading","Writing"]
w=0.3
x=np.arange(len(subs))
mpl.figure(figsize=(9,7))
mpl.bar(x-w/2,avg.loc["none"],width=w,label="None")
mpl.bar(x+w/2,avg.loc["completed"],width=w,label="Completed")
mpl.title("Test Preparation Course Analysis")
mpl.ylabel("Scores")
mpl.xlabel("Subjects")
mpl.xticks(ticks=x,labels=subs)
mpl.legend(loc="upper right")
mpl.show()
```
