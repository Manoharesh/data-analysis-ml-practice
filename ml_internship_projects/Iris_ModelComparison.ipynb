```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

Loading the iris dataset

```python
df=load_iris()
inputs=df.data
target=df.target
```

```python
# Spliting the data into train and test sets
x_train,x_test,y_train,y_test=train_test_split(inputs,target,train_size=0.8,random_state=7)
```

Initialising Models

```python
# Initialising Decision Tree Classifier model
dtcmodel=DecisionTreeClassifier()
```

```python
# Initialising KNN model
knnmodel=KNeighborsClassifier()
```

```python
# Initialising Logistic Regression model
lrmodel = LogisticRegression()
```

```python
# Initialising SVM model
svmmodel=SVC()
```

```python
# Initialising Random Forest model
rfmodel=RandomForestClassifier()
```

Training Models

```python
dtcmodel.fit(x_train,y_train)
```

```python
knnmodel.fit(x_train,y_train)
```

```python
lrmodel.fit(x_train,y_train)
```

```python
svmmodel.fit(x_train,y_train)
```

```python
rfmodel.fit(x_train,y_train)
```

Evaluating Models

```python
dtcscore=dtcmodel.score(x_test,y_test)
```

```python
knnscore=knnmodel.score(x_test,y_test)
```

```python
lrscore=lrmodel.score(x_test,y_test)
```

```python
svmscore=svmmodel.score(x_test,y_test)
```

```python
rfscore=rfmodel.score(x_test,y_test)
```

Compairing Model Perfomances

```python
model=["Decision Tree","KNN","Linear Regression","SVM","Random Forest"]
score=[round(dtcscore,2),round(knnscore,2),round(lrscore,2),round(svmscore,2),round(rfscore,2)]
plt.bar(model,score,width=0.4)
plt.title("Compairing Model Scores")
plt.xlabel("Model Names")
plt.ylabel("Scores")
for i in range(len(model)):
    plt.text(i,score[i],score[i])
plt.show()
```
